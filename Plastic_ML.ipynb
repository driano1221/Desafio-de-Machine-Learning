```python
import pandas as pd

file_path="Database Plastic Additives in Ocean 2024-02-22.xlsx"
# Load the 'Compiled Data' sheet
compiled_data_df = pd.read_excel(file_path, sheet_name='Compiled Data')



# Selecting the relevant columns from the dataset
selected_columns = [
    'Reference Year',
    'Compartment',
    'Concentration',
    '% Detected'
]

# Creating a new dataframe with the selected columns
reduced_df = compiled_data_df[selected_columns]

# Checking the data types of the selected columns
data_types = reduced_df.dtypes

```

```python
# Converting 'Concentration' and '% Detected' to numeric types, forcing errors to NaN
reduced_df['Concentration'] = pd.to_numeric(reduced_df['Concentration'], errors='coerce')
reduced_df['% Detected'] = pd.to_numeric(reduced_df['% Detected'], errors='coerce')

# Converting specified columns to categorical
reduced_df['Reference Year'] = reduced_df['Reference Year'].astype('category')
reduced_df['Compartment'] = reduced_df['Compartment'].astype('category')


# Getting the category counts and percentages
reference_year_categories = reduced_df['Reference Year'].value_counts(normalize=True) * 100
compartment_categories = reduced_df['Compartment'].value_counts(normalize=True) * 100

# Getting the number of rows in the dataset
num_rows = reduced_df.shape[0]


```

```python
# Checking the data types and the proportion of missing values
data_info = reduced_df.info()
missing_data = reduced_df.isnull().mean() * 100

# Creating a table to display the proportion of missing data
missing_data_table = pd.DataFrame({
    'Variable': missing_data.index,
    'Missing Percentage': missing_data.values
})


missing_data_table
```

Por fim, escolhemos essas 4 variáveis com as seguintes informações:

2. **Compartment**: Para entender o tipo de amostra analisada (plástico, água, sedimento ou biota).
3. **Concentration**: Para medir a quantidade de composto na amostra.

```python
# Imputing missing values for 'Concentration' using the mean
reduced_df['Concentration'].fillna(reduced_df['Concentration'].mean(), inplace=True)

# Imputing missing values for '% Detected' using the mean
reduced_df['% Detected'].fillna(reduced_df['% Detected'].mean(), inplace=True)

# Verifying the imputation
missing_data_after_imputation = reduced_df.isnull().mean() * 100
```

1. **Remoção de Outliers**: Remover os outliers pode ser uma abordagem se esses valores forem considerados ruído.
2. **Transformação de Dados**: Aplicar uma transformação, como logaritmo, pode ajudar a reduzir a influência dos outliers.
3. **Manter os Outliers**: Dependendo da natureza do estudo, pode ser necessário manter esses valores para análise mais detalhada.

```python
# Removing outliers using IQR method
Q1 = reduced_df[numeric_columns].quantile(0.25)
Q3 = reduced_df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Removing outliers
cleaned_df = reduced_df[
    (reduced_df['Concentration'] >= lower_bound['Concentration']) &
    (reduced_df['Concentration'] <= upper_bound['Concentration']) &
    (reduced_df['% Detected'] >= lower_bound['% Detected']) &
    (reduced_df['% Detected'] <= upper_bound['% Detected'])
]

# Applying logarithmic transformation to the cleaned data
log_cleaned_df = cleaned_df.copy()
log_cleaned_df['Concentration'] = np.log1p(log_cleaned_df['Concentration'])
log_cleaned_df['% Detected'] = np.log1p(log_cleaned_df['% Detected'])

# Creating box plots with logarithmic scale for the cleaned data
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

for i, col in enumerate(numeric_columns):
    sns.boxplot(data=log_cleaned_df, y=col, ax=axes[i])
    axes[i].set_title(f'Box plot of {col} (Log Transformed)')

plt.tight_layout()
plt.show()

# Creating a statistical summary for the cleaned dataset
stat_summary_cleaned = cleaned_df.describe(include='all')

stat_summary_cleaned
```

##  feature engineering

Para melhorar o nosso dataset, podemos considerar a criação de novas variáveis com base nas existentes, que podem fornecer informações adicionais ou insights que não são imediatamente aparentes nas variáveis originais. Aqui estão algumas sugestões de feature engineering que podemos realizar:

1. **Categoria de Concentração**: Podemos criar uma variável categórica baseada na "Concentration" para agrupar os dados em diferentes faixas de concentração (baixa, média, alta).

2. **Detecção Binária**: Podemos criar uma variável binária para indicar se um composto foi detectado ou não, com base na variável "% Detected".

3. **Ano Agrupado**: Agrupar os anos de referência em períodos (por exemplo, década) pode ajudar a analisar tendências temporais.

### Implementação das Novas Variáveis

Vamos criar essas novas variáveis e verificar como elas melhoram o nosso dataset.

#### 1. Categoria de Concentração
- Baixa: Concentração < 25%
- Média: Concentração entre 25% e 75%
- Alta: Concentração > 75%

#### 2. Detecção Binária
- Detectado: % Detected > 0
- Não Detectado: % Detected == 0

#### 3. Ano Agrupado
- Agrupar os anos em períodos de 5 anos.



```python
# Creating new feature 'Concentration Category'
def concentration_category(value):
    if value < log_cleaned_df['Concentration'].quantile(0.25):
        return 'Low'
    elif value < log_cleaned_df['Concentration'].quantile(0.75):
        return 'Medium'
    else:
        return 'High'

log_cleaned_df['Concentration Category'] = log_cleaned_df['Concentration'].apply(concentration_category)

# Creating new feature 'Detection Binary'
log_cleaned_df['Detection Binary'] = log_cleaned_df['% Detected'].apply(lambda x: 1 if x > 0 else 0)

# Creating new feature 'Year Group'
log_cleaned_df['Year Group'] = pd.cut(log_cleaned_df['Reference Year'], bins=list(range(1970, 2025, 5)), labels=[f"{i}-{i+4}" for i in range(1970, 2020, 5)])

# Displaying the new features
new_features_df = log_cleaned_df[['Concentration Category', 'Detection Binary', 'Year Group']]


new_features_df.head()
```


### Novas Variáveis Criadas

1. **Concentration Category**:
   - **Low**: Concentração < Percentil 25
   - **Medium**: Concentração entre Percentil 25 e 75
   - **High**: Concentração > Percentil 75

2. **Detection Binary**:
   - **1**: Detecção ( % Detected > 0 )
   - **0**: Não Detecção ( % Detected == 0 )

3. **Year Group**:
   - Agrupamento de anos em períodos de 5 anos (1970-1974, 1975-1979, ..., 2015-2019, 2020-2024)

Essas novas variáveis fornecem uma categorização mais granular e podem ajudar a identificar padrões ou tendências ocultas nos dados.

# Aplicação Machine Learning

Para aplicar técnicas de Machine Learning ao nosso dataset, considerarei duas abordagens diferentes: Regressão e Classificação. Vamos definir os problemas que vamos abordar:

1. **Regressão**: Prever a concentração de compostos plásticos ("Concentration") com base nas outras variáveis do dataset.
2. **Classificação**: Classificar se a concentração é "Alta" ou não com base nas outras variáveis do dataset.

### Passos para a Análise

1. **Preparação dos Dados**:
   - Separar os dados em conjuntos de treino e teste.
   - Normalizar os dados.
   - Converter variáveis categóricas em variáveis dummy (one-hot encoding).

2. **Aplicação das Técnicas**:
   - Regressão Linear.
   - Regressão com Árvores de Decisão.
   - Classificação com Regressão Logística.
   - Classificação com Random Forest.

3. **Avaliação das Métricas**:
   - Para Regressão: Mean Absolute Error (MAE), Mean Squared Error (MSE), R².
   - Para Classificação: Acurácia, Precision, Recall, F1-Score, AUC-ROC.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import seaborn as sns

# Preparing the data
# Selecting features and target for regression and classification
features = log_cleaned_df.drop(columns=['Concentration', 'Concentration Category'])
target_regression = log_cleaned_df['Concentration']
target_classification = log_cleaned_df['Concentration Category'].apply(lambda x: 1 if x == 'High' else 0)

# Splitting the data into training and testing sets
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(features, target_regression, test_size=0.2, random_state=42)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(features, target_classification, test_size=0.2, random_state=42)

# Defining the preprocessing steps
numeric_features = ['Reference Year', '% Detected']
categorical_features = ['Compartment', 'Detection Binary', 'Year Group']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)])

# Defining the pipelines for regression and classification
pipeline_regression_lr = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', LinearRegression())])
pipeline_regression_dt = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', DecisionTreeRegressor())])
pipeline_classification_lr = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', LogisticRegression())])
pipeline_classification_rf = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', RandomForestClassifier())])

# Training the models
pipeline_regression_lr.fit(X_train_reg, y_train_reg)
pipeline_regression_dt.fit(X_train_reg, y_train_reg)
pipeline_classification_lr.fit(X_train_clf, y_train_clf)
pipeline_classification_rf.fit(X_train_clf, y_train_clf)

# Making predictions
y_pred_reg_lr = pipeline_regression_lr.predict(X_test_reg)
y_pred_reg_dt = pipeline_regression_dt.predict(X_test_reg)
y_pred_clf_lr = pipeline_classification_lr.predict(X_test_clf)
y_pred_clf_rf = pipeline_classification_rf.predict(X_test_clf)

# Calculating metrics for regression
mae_lr = mean_absolute_error(y_test_reg, y_pred_reg_lr)
mse_lr = mean_squared_error(y_test_reg, y_pred_reg_lr)
r2_lr = r2_score(y_test_reg, y_pred_reg_lr)

mae_dt = mean_absolute_error(y_test_reg, y_pred_reg_dt)
mse_dt = mean_squared_error(y_test_reg, y_pred_reg_dt)
r2_dt = r2_score(y_test_reg, y_pred_reg_dt)

# Calculating metrics for classification
accuracy_lr = accuracy_score(y_test_clf, y_pred_clf_lr)
precision_lr = precision_score(y_test_clf, y_pred_clf_lr)
recall_lr = recall_score(y_test_clf, y_pred_clf_lr)
f1_lr = f1_score(y_test_clf, y_pred_clf_lr)
roc_auc_lr = roc_auc_score(y_test_clf, y_pred_clf_lr)

accuracy_rf = accuracy_score(y_test_clf, y_pred_clf_rf)
precision_rf = precision_score(y_test_clf, y_pred_clf_rf)
recall_rf = recall_score(y_test_clf, y_pred_clf_rf)
f1_rf = f1_score(y_test_clf, y_pred_clf_rf)
roc_auc_rf = roc_auc_score(y_test_clf, y_pred_clf_rf)

# Creating a DataFrame to display the metrics
metrics_regression = pd.DataFrame({
    'Model': ['Linear Regression', 'Decision Tree'],
    'MAE': [mae_lr, mae_dt],
    'MSE': [mse_lr, mse_dt],
    'R2': [r2_lr, r2_dt]
})

metrics_classification = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest'],
    'Accuracy': [accuracy_lr, accuracy_rf],
    'Precision': [precision_lr, precision_rf],
    'Recall': [recall_lr, recall_rf],
    'F1-Score': [f1_lr, f1_rf],
    'ROC AUC': [roc_auc_lr, roc_auc_rf]
})


metrics_regression, metrics_classification
```
